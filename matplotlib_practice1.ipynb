```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
xpoints = np.array([0,6])
xpoints
```

```python
ypoints = np.array([0,250])
```

```python
# Plot a line plot
plt.plot(xpoints, ypoints)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Plotting without line
plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
# Plotting multiple points
xpoints = np.array([1,3,6,8])
ypoints = np.array([6,8,4,12])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
# Default x points: When we don't indicate x points, the x will get values: 0,1,2,3...
ypoints = np.array([3,4,5,1,5,6,8])
plt.plot(ypoints)
plt.show()
```

```python
# Use markers
plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
plt.plot(ypoints,'o')
plt.show()
```

```python
# Star markers
plt.plot(ypoints, marker='*')
plt.show()
```

```python
# Square markers
plt.plot(ypoints, marker='s')
```

```python
# Diamond markers
plt.plot(ypoints,marker='D')
```

```python
ypoints = np.array([1,8,3,10])
plt.plot(ypoints, 'o:r')
```

```python
# Adjust marker size and color of the edge
plt.plot(ypoints, marker='s', ms=20, mec='r')
```

```python
# Change the color of the face of markers
plt.plot(ypoints, marker='d', ms=10, mfc='g')
```

```python
# Change both the colors of the edge and the face of markers
plt.plot(ypoints, marker='*', ms=15, mec='r', mfc='r')
```

```python
# Change the color to 'hot pink'
plt.plot(ypoints, marker='*', ms=15, mec='hotpink', mfc='hotpink')
```
